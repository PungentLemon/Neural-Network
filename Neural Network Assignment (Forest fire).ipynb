## Tuning of Hyperparameters

#### Batch Size and Epochs

```python
from sklearn.model_selection import GridSearchCV,KFold
from keras.layers import Dense
from scikeras.wrappers import KerasClassifier
from keras.optimizer_v2.adam import Adam
```

```python
# Creating model function

def modelBE():
    model= Sequential()
    model.add(Dense(12,input_dim=28, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(8,kernel_initializer='uniform', activation='relu'))
    model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
    
    adam= Adam(learning_rate=0.01)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['acc'])
    return model
```

```python
#Creating model
model_B_E= KerasClassifier(model=modelBE, verbose=2)

#Define The grid search parameters
batch_size=[10,20,40]
epochs= [10,20,30]

#Defining grid search dictionary
param_B_E= dict(batch_size=batch_size, epochs=epochs)

#Building and fitting GridSearchCV

grid_B_E= GridSearchCV(estimator=model_B_E,param_grid=param_B_E, cv=KFold(),verbose=1)
```

```python
result_B_E= grid_B_E.fit(X,y)
```

```python
#Summary
print('Best score is {} for param {}'.format(result_B_E.best_score_, result_B_E.best_params_))
```

#### Learning rate and Dropout rate

```python
from keras.layers import Dropout

#Defining grid search param
learning_rate=[0.001,0.01,0.1]
dropout_rate=[0.0,0.1,0.2]

#Creating grid search dictionary
param_D_L= dict(learning_rate=learning_rate, dropout_rate=dropout_rate)

#Creating the model function
def model_DL(learning_rate,dropout_rate):
    model=Sequential()
    model.add(Dense(8,input_dim=28,kernel_initializer='normal',activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(4,input_dim=8,kernel_initializer='normal', activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1,activation='sigmoid'))
    
    adam = Adam(learning_rate=learning_rate)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['acc'])
    return model

#Create the model
model_D_L =KerasClassifier(model=model_DL, verbose=0, batch_size=10,epochs=30, dropout_rate=dropout_rate, learning_rate=learning_rate)
```

```python
#Buiding and fitting the GridSearchCV

grid_D_L= GridSearchCV(estimator=model_D_L, param_grid=param_D_L, cv=KFold(),verbose=1)
result_D_L= grid_D_L.fit(X,y)
```

```python
print('Best score:{}, best param{}'.format(result_D_L.best_score_, result_D_L.best_params_))
```

#### Activation Function and Kernel Initializer

```python
# Defining the grid search parameters

activation_function=['softmax','relu','tanh','linear']
kernel= ['uniform','normal','zero']

#creating param dictionary

param_A_K = dict(activation_function=activation_function, kernel=kernel)

#Creating model function
def modelAK(activation_function,kernel):
    model=Sequential()
    model.add(Dense(12,input_dim=28,kernel_initializer=kernel,activation=activation_function))
    model.add(Dropout(0.2))
    model.add(Dense(8,input_dim=12, kernel_initializer=kernel, activation=activation_function))
    model.add(Dropout(0.2))
    model.add(Dense(1,activation='sigmoid'))
    
    adam= Adam(learning_rate=0.01)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['acc'])
    return model

#Creating model
model_A_K= KerasClassifier(model=modelAK,verbose=1,batch_size=10, epochs=30, activation_function=activation_function,kernel=kernel)

#GridSearch
grid_A_K= GridSearchCV(estimator=model_A_K,param_grid=param_A_K, cv=KFold(),verbose=1)
result_A_K=grid_A_K.fit(X,y)
```

```python
print('Best score is {} and best param are {}'.format(result_A_K.best_score_,result_A_K.best_params_))
```

#### Number of neurons in activation function

```python
#Defining the grid search param
neuron1=[12,16,18]
neuron2=[8,12,16]

#param dictionary
param_neuron=dict(neuron1=neuron1,neuron2=neuron2)

#Creating model function
def model_neu(neuron1,neuron2):
    model=Sequential()
    model.add(Dense(neuron1, input_dim=28, kernel_initializer='normal', activation='linear'))
    model.add(Dropout(0.2))
    model.add(Dense(neuron2, input_dim=neuron1, kernel_initializer='normal', activation='linear'))
    model.add(Dropout(0.2))
    model.add(Dense(1,activation='sigmoid'))
    
    adam= Adam(learning_rate=0.01)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['acc'])
    return model
```

```python
#creating model
model_neuron = KerasClassifier(model=model_neu,verbose=1, batch_size=10, epochs=30, neuron1=neuron1, neuron2=neuron2)

#GridsearchCV
grid_neuron= GridSearchCV(estimator=model_neuron, param_grid=param_neuron,cv=KFold(),verbose=1)
result_neuron=grid_neuron.fit(X,y)
```

```python
print('Best score is {} and best param are {}'.format(result_neuron.best_score_,result_neuron.best_params_))
```
