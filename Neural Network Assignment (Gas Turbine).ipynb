#### Tuning of Hyperparameters :- Batch Size and Epochs

```python
#Importing the necessary packages
from sklearn.model_selection import GridSearchCV, KFold
from keras.models import Sequential #to make sure the proccess is carried in one direction direction 
from keras.layers import Dense #this creates layers for NN
from scikeras.wrappers import KerasRegressor
from keras.optimizer_v2.adam import Adam
```

```python
# create model

### input_dim for 1st layer= no of independent/ 'x' features 

def model_b_e():
    model = Sequential()
    model.add(Dense(12, input_dim=3, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

    adam=Adam(learning_rate=0.01)
    model.compile(loss='mean_squared_error', optimizer=adam, metrics=['mse'])
    return model
```

```python
# Create the model
model_batch_epoc = KerasRegressor(model = model_b_e,verbose = 1)

# Define the grid search parameters
batch_size = [10,20,40]
epochs = [5,10,15]
# Make a dictionary of the grid search parameters
param_grid = dict(batch_size = batch_size,epochs = epochs)
# Build and fit the GridSearchCV
grid_b_e = GridSearchCV(estimator = model_batch_epoc,param_grid = param_grid,cv = KFold(),verbose =10)
resut_b_e = grid_b_e.fit(X,y)
```

```python
print('Best score is : {}, using {}\n'.format(resut_b_e.best_score_,resut_b_e.best_params_))
means = resut_b_e.cv_results_['mean_test_score']
stds = resut_b_e.cv_results_['std_test_score']
params = resut_b_e.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
  print('mean is{} and std is{} for {}\n'.format(mean, stdev, param))
```

#### Tuning of Hyperparameters:- Learning rate and Drop out rate


```python
from keras.layers import Dropout

learning_rate =[0.001, 0.01, 0.1]
dropout_rate = [0.0, 0.1, 0.2]

#Defining the model function

def model_L_D(learning_rate,dropout_rate):
    model=Sequential()
    model.add(Dense(8, input_dim=3, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(4, input_dim=3, kernel_initializer='normal', activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1,activation='sigmoid'))
    
    adam= Adam(learning_rate=learning_rate)
    model.compile(loss='mean_squared_error', optimizer=adam, metrics=['mse'])
    return model
```

```python
#Creating the model
model_Le_Dr=KerasRegressor(model = model_L_D, verbose=1, batch_size=10, epochs=10, dropout_rate=dropout_rate, learning_rate=learning_rate)
```

```python
# Making a dictionary of the grid search param
param_D_L = dict(learning_rate=learning_rate, dropout_rate=dropout_rate)

#Building and fitting the GridSearchCV

grid_D_L =GridSearchCV(estimator=model_Le_Dr,param_grid=param_D_L, cv=KFold(), verbose=1)
result_D_L = grid_D_L.fit(X,y)
```

```python
# Summary

print('Best score is {}, with param: {}\n'. format(result_D_L.best_score_,
                                                result_D_L.best_params_))
mean= result_D_L.cv_results_['mean_test_score']
std= result_D_L.cv_results_['std_test_score']
param= result_D_L.cv_results_['params']
for m,s,p in zip(mean,std,param):
    print('Mean is {} and std is {} for param:{}\n'. format(m,s,p))
```

#### Tuning of Hyperparameters:- Activation Function and Kernel Initializer

```python
# Defining the grid seach param
activation_func = ['softmax', 'relu', 'tanh', 'linear']
kernel =['uniform', 'normal', 'zero']

# Creating model function

def model_A_K(activation_func,kernel):
    model =Sequential()
    model.add(Dense(8,input_dim=3, kernel_initializer = kernel, activation= activation_func))
    model.add(Dropout(0.0))
    model.add(Dense(4, input_dim=3, kernel_initializer=kernel, activation=activation_func))
    model.add(Dropout(0.0))
    model.add(Dense(1,activation='sigmoid'))
    
    adam = Adam(learning_rate=0.001)
    model.compile(loss='mean_squared_error', optimizer=adam, metrics=['mse'])
    return model

#create the model
model_Ac_Ke= KerasRegressor(model=model_A_K,verbose=1, batch_size=10, epochs=10, activation_func=activation_func, kernel=kernel)

#make a dictionary of the grid search param
param_A_K = dict(activation_func=activation_func, kernel=kernel)

#Building and fitting GridSearchCV
grid_A_K = GridSearchCV(estimator=model_Ac_Ke,param_grid=param_A_K,cv=KFold(),verbose=1)
result_A_K = grid_A_K.fit(X,y)
```

```python
# Summary

print('Best score is {}, with param: {}\n'. format(result_A_K.best_score_,
                                                result_A_K.best_params_))
mean= result_A_K.cv_results_['mean_test_score']
std= result_A_K.cv_results_['std_test_score']
param= result_A_K.cv_results_['params']
for m,s,p in zip(mean,std,param):
    print('Mean is {} and std is {} for param:{}\n'. format(m,s,p))
```

#### Tuning of the Hyperparameter:- Number of Neurons in the activation layer

```python
#Defining the grid search param
neuron1 =[4,8,12]
neuron2=[2,4,8]

#Defining the model function
def model_n(neuron1,neuron2):
    model= Sequential()
    model.add(Dense(neuron1,input_dim=3,kernel_initializer='uniform', activation='relu'))
    model.add(Dropout(0.0))
    model.add(Dense(neuron2,input_dim=neuron1, kernel_initializer='uniform', activation='relu'))
    model.add(Dropout(0.0))
    model.add(Dense(1, activation='sigmoid'))
    
    adam=Adam(learning_rate=0.001)
    model.compile(loss='mean_squared_error', optimizer=adam , metrics=['mse'])
    return model

#Create the model
model_neuron =KerasRegressor(model=model_n, verbose=1, neuron1=neuron1, neuron2=neuron2)

#Creating param dictionary
param_neuron= dict(neuron1=neuron1,neuron2=neuron2)

#Building and fitting GridSearchCV
grid_neu= GridSearchCV(estimator=model_neuron, param_grid=param_neuron,cv=KFold(),verbose=2)
result_neu= grid_neu.fit(X,y)
```

```python
# Summary

print('Best score is {}, with param: {}\n'. format(result_neu.best_score_,
                                                result_neu.best_params_))
mean= result_neu.cv_results_['mean_test_score']
std= result_neu.cv_results_['std_test_score']
param= result_neu.cv_results_['params']
for m,s,p in zip(mean,std,param):
    print('Mean is {} and std is {} for param:{}\n'. format(m,s,p))
```
